```python
######### get the data to examine and use
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
books = pd.read_csv(f"{path}/books.csv")
books
```

```python
############# find the patterns in the missing data to see if there is a bias
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
# from the heatmap, avg rating, num pages, and ratings cnt are all associated with each other - possible bias (maybe came from another source)
```

```python
#########################################
# look at fields with "description" missing (desc = core field)
# create numeric representation of which books are missing

import numpy as np
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2026 - books["published_year"]
```

```python
columns_of_interest = ["num_pages","age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method= "spearman") # spearman correlation coefficient is a better fit for non cont values than pearson

# convert correlation matrix in to seaborn heatmap
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})

heatmap.set_title("Correlation Heatmap")
plt.show() # shows that there is no big correlation with a missing description to other data
```

```python
############ remove the observations (books) with missing values
# find how many books have missing values
# books[(books["description"].isna()) |
#       (books["num_pages"].isna()) |
#       (books["published_year"].isna()) |
#       (books["average_rating"].isna())
# ] # these results are just under 5% of the entire data set so we can remove

# then negate it to show the ones that do not contain the missing observations
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
] #yields a dataframe with complete
book_missing
```

```python
############## Categories Field - has a lot of distinct values that isn't usable for us

# look at the distribution of categories by taking value of counts of each category
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False) #click chart view and setting to view as chart

# there is a huge tail of books with a distinct category here, we will clean up more later and standarize it
```

```python
####### Descriptions refinement - important because the description needs to be meaningful enough to recommend it
book_missing
```

```python
# check how long desc are - usually < 3 words is useless
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
book_missing #look as this as a histogram (x, y: words in desc, y is sorted by count)
```

```python
# here in the histogram, there is no natural breakpoint between really short desc and meaningful desc - so we need to create an arbitrary cut point ourselves

# start with the smaller descriptions:
book_missing.loc[book_missing["words_in_description"].between(1,4), "description"] #look for desc with 1-4 words and return description
```

```python
# check for slightly larger descriptions to see if this range of words is meaningful:
book_missing.loc[book_missing["words_in_description"].between(5,14), "description"] #can conclude that this range still isn't very helpful
```

```python
# try the next size range and see if desc is meaningful - YES (ish)! desc is starting to be informative now, but not enough for a user to understand
book_missing.loc[book_missing["words_in_description"].between(15,24), "description"]
```

```python
# try next size -> YES! now at a point whee desc gives us enough information for program to be able to recommend users the book
book_missing.loc[book_missing["words_in_description"].between(25,34), "description"]
```

```python
# now that we determined 25 words is a good cut off for the desc, lets clean up our data set and keep only those 25 or more words for desc
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
book_missing_25_words
```

```python
# Final part of data cleaning: decide what to do with subtitle field (rn unusable on its own due to so many empty values)
# many book naming conventions include title + subtitle so create new variable to conjoin them

# when subtitle is missing, just use the title - if subtitle is there, combine with title
book_missing_25_words = book_missing_25_words.copy()

book_missing_25_words["title_and_subtitle"] = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],
    book_missing_25_words["title"].astype(str)
    + ": "
    + book_missing_25_words["subtitle"].astype(str),
)

book_missing_25_words
```

```python
# combine the unique identifier isbn13 with description for later vector usage:
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)

book_missing_25_words
```

```python
# drop the columns that we don't need - either used for clean up/filtering and save file
(
    book_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1).to_csv("books_cleaned_up", index=False)
)
```
